```python
from scipy.sparse import diags
import matplotlib.pyplot as plt
Nx,Ny = 50,50
Nc = Nx * Ny

e = np.ones(Nc)
print(e)

A = diags([e,e,-4*e,e,e], [-Nx, -1, 0, 1, Nx], shape=(Nc,Nc))

plt.spy(A, marker='o',markersize=6)
plt.tight_layout()
```

```python
print(A)
```

```python
print(np.arange(Nc).reshape(Nx,Ny)[::-1])
```

```python
T_bottom = np.arange(Nx)
print(T_bottom)
```

```python
T_top = T_bottom + (Ny-1)*Nx
print(T_top)
```

```python
T_left = np.arange(Ny)*(Nx)
print(T_left)
```

```python
T_right = T_left + Nx-1
print(T_right)
```

```python
bnd_all = np.unique(np.concatenate((T_bottom,T_top,T_left,T_right)))
print(bnd_all)
```

```python
from scipy.sparse import lil_matrix

A = lil_matrix(A)
A[bnd_all,:] = 0
A[bnd_all,bnd_all] = 1
plt.spy(A,marker='o',markersize=5)
```

```python
b = np.zeros(Nc)
b[T_bottom] = 10
b[T_top] = 20
b[T_left] = 30
b[T_right] = 40
plt.spy(b[:,None],marker='o')
```

```python
from scipy.sparse.linalg import spsolve
T = spsolve(A.tocsc(),b)
print(T.reshape(Nx,Ny)[::-1])
```

```python
x,y = np.meshgrid(np.linspace(0,1,Nx), np.linspace(0,1,Ny))
fig, ax = plt.subplots(subplot_kw={"projection":"3d"})
ax.plot_surface(x,y,T.reshape(Nx,Ny))
```

## Using the functions in the laplace_demo.py files

```python
%reset -f
```

```python
from laplace_demo import create_laplace_coefficient_matrix, set_boundary_conditions
from scipy.sparse.linalg import spsolve
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm


Nx, Ny = 20, 20
Tb = {'bottom': 10, 'top': 20, 'left': 200, 'right': 500}

A,b = create_laplace_coefficient_matrix(Nx,Ny)
A,b = set_boundary_conditions(A,b,Tb,Nx,Ny)

T = spsolve(A,b).reshape(Nx,Ny)

x,y = np.meshgrid(np.linspace(0,1,Nx), np.linspace(0,1,Ny))
fig, ax = plt.subplots(subplot_kw={"projection":"3d"})
surf = ax.plot_surface(x,y,T.reshape(Nx,Ny),cmap=cm.inferno,edgecolor='black')
fig.colorbar(surf,shrink=0.5)
plt.show()
```

## Computing and comparing the exact solution

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from scipy.sparse.linalg import spsolve

Nx, Ny = 20, 20
Tb = {'bottom': 0, 'top': 1, 'left': 0, 'right': 0}

A,b = create_laplace_coefficient_matrix(Nx,Ny)
A,b = set_boundary_conditions(A,b,Tb,Nx,Ny)

T = spsolve(A,b).reshape(Nx,Ny)

x,y = np.meshgrid(np.linspace(0,1,Nx), np.linspace(0,1,Ny))

T_ex = np.zeros_like(x)

for n in range(1,100):
    m = 2*n - 1
    term = np.sin(m*np.pi*x) * np.sinh(m*np.pi*y)/(m * np.sinh(m*np.pi))
    T_ex += term

T_ex *= (4/np.pi)

fig, ax = plt.subplots(subplot_kw={"projection":"3d"})
surf = ax.plot_surface(x,y,T - T_ex,cmap=cm.inferno,edgecolor='black')
fig.colorbar(surf,shrink=0.5)
plt.show()
```
